```python
import matplotlib.pyplot as plt 
```

```python
# data to display on plots 
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 
```

```python
# This will plot a simple bar chart
plt.bar(x, y)
```

```python
# Title to the plot
plt.title("Bar Chart")
```

```python
# Adding the legends
plt.legend(["bar"])
plt.show()
```
