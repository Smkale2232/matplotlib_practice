```python
# importing matplotlib
import matplotlib.pyplot as plt 
```

```python
# making a simple plot
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
```

```python
# creating error
y_error = 0.2
```

```python
# plotting graph
plt.plot(x, y)
```

```python
plt.errorbar(x, y,
            yerr = y_error,
            fmt ='o')
```
