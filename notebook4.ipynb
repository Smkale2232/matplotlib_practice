```python
import matplotlib.pyplot as plt
```

```python
# List of Days
days = [1, 2, 3, 4, 5]
```

```python
# No of Study Hours
Studying = [7, 8, 6, 11, 7]
```

```python
# No of Playing Hours
playing = [8, 5, 7, 8, 13]
```

```python
# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
            colors =['r', 'c'])
```

```python
# Days
plt.xlabel('Days')
```

```python
# No of hours
plt.ylabel('No of Hours')
```

```python
# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')
```

```python
# Displaying Graph
plt.show()
```
